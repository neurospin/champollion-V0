```python
import pandas as pd
import glob
```

```python
volumes_file = "/neurospin/dico/data/bv_databases/human/automatically_labeled/synesthetes/*/morphometry/brain_volumes.csv"
participants_file = "/neurospin/dico/data/deep_folding/current/datasets/synesthetes/participants.csv"
```

```python
syn_files = glob.glob(volumes_file)
syn_list = []
for syn_file in syn_files:
    syn_list.append(pd.read_csv(syn_file, sep=";", index_col=0)[["both.brain_volume"]])
syn = pd.concat(syn_list, axis=0)
syn = syn.rename_axis("participant_id")
syn = syn.rename(columns={"both.brain_volume": "brain_volume"})
syn.head()
```

```python
participants = pd.read_csv(participants_file, index_col=0)
participants.head()
```

```python
participants_out = syn.merge(participants, left_index=True, right_index=True)
```

```python
participants_out
```

```python
participants_out.to_csv(participants_file)
```
