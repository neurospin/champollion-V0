```python
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_summary = "/neurospin/dico/data/deep_folding/current/models/Champollion_V0/summary"
```

```python
glob.glob(f"{path_summary}/*")
```

```python
files = ['/neurospin/dico/data/deep_folding/current/models/Champollion_V0/summary/SexSmallBrain_ukb_without_residualization.csv',
         '/neurospin/dico/data/deep_folding/current/models/Champollion_V0/summary/isBigBrain_ukb_sex_balanced.csv',
         '/neurospin/dico/data/deep_folding/current/models/Champollion_V0/summary/isOld_ukb_without_residualization.csv',
         '/neurospin/dico/data/deep_folding/current/models/Champollion_V0/summary/site_hcp_ukb_younger_without_residualization.csv']
names = ['Sex', 'BigBrain', 'Age', 'site']
```

```python
list_df = []
for f, name in zip(files, names):
    df = pd.read_csv(f, index_col=0)[["region", "roc_auc"]]
    df = df.groupby(['region']).mean()
    df = df.rename(columns={'roc_auc': name})
    list_df.append(df)
result = pd.concat(list_df, axis=1)
```

```python
pd.plotting.scatter_matrix(result, alpha=1)
```
