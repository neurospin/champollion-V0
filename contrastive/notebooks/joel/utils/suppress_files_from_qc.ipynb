```python
import glob
import os
import pandas as pd
```

```python
path = "/neurospin/dico/data/deep_folding/current/datasets/UkBioBank40"
side = "L"
qc_file = "/neurospin/dico/data/deep_folding/current/datasets/UkBioBank40/qc.tsv"
```

```python
qc = pd.read_csv(qc_file, index_col=0, sep='\t')
qc.head()
```

```python
def remove_files(subpath, prefix, subject, postfix=".nii.gz"):
    path_files = glob.glob(f"{path}/{subpath}/{side}/{side}{prefix}{subject}{postfix}")
    for path_file in path_files:
        path_file_minf = f"{path_file}.minf"
        if os.path.exists(path_file):
            os.remove(path_file)
            print(f'{subject}..{subpath} removed..')
        if os.path.exists(path_file_minf):
            os.remove(path_file_minf)
```

```python
def remove_files_before_masking(subpath, prefix, subject):
    path_files = glob.glob(f"{path}/{subpath}/{side}_before_masking/{side}{prefix}{subject}.nii.gz")
    for path_file in path_files:
        path_file_minf = f"{path_file}.minf"
        if os.path.exists(path_file):
            os.remove(path_file)
            print(f'{subject}..{subpath} removed..')
        if os.path.exists(path_file_minf):
            os.remove(path_file_minf)
```

```python
def remove_crops(subpath, postfix, subject):
    path_files = glob.glob(f"{path}/{subpath}/{subject}{postfix}.nii.gz")
    for path_file in path_files:
        path_file_minf = f"{path_file}.minf"
        if os.path.exists(path_file):
            os.remove(path_file)
            print(f'{subject}..{subpath} removed..')
        if os.path.exists(path_file_minf):
            os.remove(path_file_minf)
```

```python

for subject in qc.index:
    qc_value = qc.loc[subject, "qc"]
    subject_ = f"{subject}_"
    subject = f"{subject}"
    if qc_value == 0:
        print(subject)
        remove_files("skeletons/raw", "skeleton_generated_", subject)
        remove_files("skeletons/2mm", "resampled_skeleton_", subject)
        remove_files("extremities/raw", "extremities_", subject)
        remove_files("extremities/2mm", "resampled_extremities_", subject)
        remove_files("foldlabels/raw", "foldlabel_", subject)
        remove_files("foldlabels/2mm", "resampled_foldlabel_", subject)
        remove_files("transforms", "transform_to_ICBM2009c_", subject, ".trm")
        remove_files_before_masking("foldlabels/2mm", "resampled_foldlabel_", subject)
        remove_files_before_masking("extremities/2mm", "resampled_extremities_", subject)
        remove_crops(f"crops/2mm/F.I.P./mask/{side}labels", "_cropped_foldlabel", subject)
        remove_crops(f"crops/2mm/F.I.P./mask/{side}crops", "_cropped_skeleton", subject)
        remove_crops(f"crops/2mm/F.I.P./mask/{side}extremities", "_cropped_extremities", subject)
        remove_crops(f"crops/2mm/F.I.P./mask/{side}distbottom", "_cropped_distbottom", subject)
```
