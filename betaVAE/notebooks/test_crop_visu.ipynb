Visualize crops to see if well positioned

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
#sulcus='CINGULATE.baby'
sulcus='S.C.'
side='L'
skel_dir = f'/neurospin/dico/data/deep_folding/current/datasets/UkBioBank/crops/2mm/{sulcus}/mask/'
```

```python
skels = np.load(skel_dir+f'{side}skeleton.npy')
```

```python
idxs = [rd.randint(0, len(skels)) for k in range(10)]
```

```python
fig, axs = plt.subplots(ncols=len(idxs), figsize=(5*len(idxs),5))
for i, idx in enumerate(idxs):
    skel = skels[idx] > 0
    axs[i].imshow(skel[15,:,:], cmap='gray')
plt.show()
```
