```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import yaml
from betaVAE.configs.config import Config
```

# Visualisation of reconstructed slices

```python
def show_slices(inputs, outputs, n_views):
    n = 0
    while n < n_views:
        i = np.random.randint(0,inputs.shape[1])
        j = np.random.randint(7,14)

        inp = inputs[0,i,j,:,:]
        out = outputs[0,i,j,:,:]

        if len(np.unique(inp)) != 1: # not only zeros
            plt.figure(figsize=(16,6))
            plt.subplot(1,2,1)
            sns.heatmap(inp)
            plt.title(f"Input {i} slice {j}")
            plt.subplot(1,2,2)
            sns.heatmap(out)
            plt.title(f"Output {i} slice {j}")
            n += 1
```

```python
# Slices obtained with reduction == 'mean'

inputs = np.load("/neurospin/dico/agaudin/Runs/09_new_repo/Output/2023-04-03/16-16-25/input.npy")
outputs = np.load("/neurospin/dico/agaudin/Runs/09_new_repo/Output/2023-04-03/16-16-25/output.npy")
n_views = 8

show_slices(inputs, outputs, n_views)
```

```python
# Slices obtained with reduction == 'sum'

inputs = np.load("/neurospin/dico/agaudin/Runs/09_new_repo/Output/2023-04-03/15-37-55/input.npy")
outputs = np.load("/neurospin/dico/agaudin/Runs/09_new_repo/Output/2023-04-03/15-37-55/output.npy")
n_views = 8

show_slices(inputs, outputs, n_views)
```
